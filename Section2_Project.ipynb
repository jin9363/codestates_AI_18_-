```python
target = '10년후거주의사_인천시'

x = df.drop(target, axis=1)
y = df[target]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, stratify=y_train)

print('x', x_train.shape, x_val.shape, x_test.shape)
print('y', y_train.shape, y_val.shape, y_test.shape)
```

```python
scaler = MinMaxScaler()
x_train_s = scaler.fit_transform(x_train)
x_val_s = scaler.transform(x_val)
x_test_s = scaler.transform(x_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate, cross_val_predict
from sklearn.ensemble import RandomForestClassifier, BaggingRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import *
```

```python
# 최빈값 기준모델 생성

base = [y_train.mode()[0]] * len(y_train)

# 기준모델 스코어
base_accuracy = accuracy_score(y_train, base).round(3)
base_f1 = f1_score(y_train, base).round(3)
base_roc_auc = roc_auc_score(y_train, base).round(3)

print(f'기준모델의 정확도는 {base_accuracy.round(3)} 입니다')
```

```python
# RandomForest model 생성(기본)

model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(x_train_s, y_train)

train_pred_rf = model_rf.predict(x_train_s)

train_rf_accuracy = accuracy_score(y_train, train_pred_rf).round(3)

pred_rf = model_rf.predict(x_val_s)
print(confusion_matrix(y_val, pred_rf))
print(f'rf모델의 train 정확도는 {train_rf_accuracy.round(3)} 입니다')
print(classification_report(y_val, pred_rf))
```

```python
# Logistic model 생성(기본)

model_lr = LogisticRegression()
model_lr.fit(x_train_s, y_train)

train_pred_lr = model_lr.predict(x_train_s)

train_lr_accuracy = accuracy_score(y_train, train_pred_lr).round(3)

pred_lr = model_lr.predict(x_val_s)
print(confusion_matrix(y_val, pred_lr))
print(f'lr모델의 train 정확도는 {train_lr_accuracy.round(3)} 입니다')
print(classification_report(y_val, pred_lr))
```
